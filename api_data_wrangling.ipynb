```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.ffill(inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (°C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.ffill(inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Task 1: Fetch Data

```python
# Set Up Parameters

# Location coordinates for London
latitude_ldn = 51.5074
longitude_ldn = -0.1278

# Date range
start_date = '2022-01-01'
end_date = '2022-12-31'  # One year of data

# Variables to fetch
hourly_variables = ['precipitation']
```

```python
# Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

# London parameters
params_ldn = {
    'latitude': latitude_ldn,
    'longitude': longitude_ldn,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'Europe/London'
}

response_ldn = requests.get(base_url, params=params_ldn)

# Check if the London request was successful
if response_ldn.status_code == 200:
    print('London data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response_ldn.status_code}')

# Location coordinates for Seattle
latitude_sea = 47.6062
longitude_sea = -122.3321

# Seattle parameters
params_sea = {
    'latitude': latitude_sea,
    'longitude': longitude_sea,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_sea = requests.get(base_url, params=params_sea)

# Check if the Seattle request was successful
if response_sea.status_code == 200:
    print('Seattle data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response_sea.status_code}')
```

```python
# Load Data into pandas DataFrame

data_ldn = response_ldn.json()
data_sea = response_sea.json()

hourly_data_ldn = data_ldn['hourly']
hourly_data_sea = data_sea['hourly']

# Create DataFrames for London and Seattle
df_ldn = pd.DataFrame(hourly_data_ldn)
df_sea = pd.DataFrame(hourly_data_sea)

print(df_ldn.head())
print(df_sea.head())
```

## Task 2: Data Cleaning

```python
# Data Cleaning

# Convert 'time' column to datetime
df_ldn['time'] = pd.to_datetime(df_ldn['time'])
df_sea['time'] = pd.to_datetime(df_sea['time'])

# Set 'time' as the index
df_ldn.set_index('time', inplace=True)
df_sea.set_index('time', inplace=True)

# Check for missing values
df_ldn.isnull().sum()
df_sea.isnull().sum()
```

## Task 3: Resample Data

```python
# Resample the hourly data to weekly totals
weekly_precip_ldn = df_ldn.resample('W').sum()
weekly_precip_sea = df_sea.resample('W').sum()
```

## Task 4: Visualization

```python
# Combine Weekly Precipitation dataframes
combined_precip = pd.DataFrame({
    'London': weekly_precip_ldn['precipitation'],
    'Seattle': weekly_precip_sea['precipitation']
})

# Plot Comparison of Weekly precipitation totals
combined_precip.plot(kind='line', figsize=(14, 8))
plt.title('Weekly Precipitation Totals: London vs Seattle', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Combine Cumulative Precipitation dataframes
combined_precip_cumsum = pd.DataFrame({
    'London': weekly_precip_ldn['precipitation'].cumsum(),
    'Seattle': weekly_precip_sea['precipitation'].cumsum()
})

# Plot Cumulative Precipitation
combined_precip_cumsum.plot(figsize=(14, 8))
plt.title('Cumulative Precipitation in 2022: London vs Seattle', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.xticks(rotation=45)
plt.show()
```

## Task 5: Analysis

```python
total_ldn = int(weekly_precip_ldn['precipitation'].sum())
total_sea = int(weekly_precip_sea['precipitation'].sum())
print('2022 Precipitation Totals')
print(f'London: {total_ldn} mm')
print(f'Seattle: {total_sea} mm')
# Calculate the difference in total precipitation
total_dif = abs(total_ldn - total_sea)
print(f'Difference: {total_dif} mm')
```

Seattle received significantly more precipitation than London in the spring and winter months. Both cities saw drier summer months with precipitation increasing in the fall. Overall Seattle received ~469mm more precipitation than London in 2022.
